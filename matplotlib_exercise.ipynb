## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv('data/hist_data.csv')
plt.hist(df['aデータ'],bins=30,color='gold',alpha=0.7,label='aデータ')
plt.hist(df['bデータ'],bins=30,color='paleturquoise',alpha=0.7,label='bデータ')
plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_ = pd.read_csv('data/scatter_data.csv')
df0 = df_[df_['class']==0]
df1 = df_[df_['class']==1]
df2 = df_[df_['class']==2]
plt.scatter(df0['length (cm)'],df0['width (cm)'],color='red',label='class0')
plt.scatter(df1['length (cm)'],df1['width (cm)'],color='blue',label='class1')
plt.scatter(df2['length (cm)'],df2['width (cm)'],color='green',label='class2')
plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.grid()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df = pd.read_csv('data/line_data.csv')
plt.plot(df['期間'], df['件数'], marker='o', linestyle='-', color='steelblue')
plt.title('期間ごとの件数の推移')
plt.xlabel('期間')
plt.ylabel('件数')
plt.grid()
plt.xticks(rotation=30)
```

```python
plt.bar(df['期間'],df['費用'],color='steelblue')
plt.title('期間ごとの費用の推移')
plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.grid()
plt.xticks(rotation=30)
```
